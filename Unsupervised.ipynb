## Melihat Kadar Alkohol

```python
# Menghitung rata-rata kadar alkohol berdasarkan kualitas
average_alcohol = gabungan_wine_df.groupby('quality')['alcohol'].mean().reset_index()

print(average_alcohol)
```

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='quality', y='alcohol', data=average_alcohol, palette='Set2')
plt.title('Rata-rata Kadar Alkohol Berdasarkan Kualitas')
plt.xlabel('Kualitas')
plt.ylabel('Rata-rata Kadar Alkohol')

# Menambahkan label di atas setiap batang
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Menghitung rata-rata kadar pH berdasarkan kualitas
average_pH = gabungan_wine_df.groupby('quality')['pH'].mean().reset_index()

print(average_pH)
```

## Rata-rata Kadar PH di Setiap Kualitas

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='quality', y='pH', data=average_pH, palette='Set2')
plt.title('Rata-rata Kadar pH Berdasarkan Kualitas')
plt.xlabel('Kualitas')
plt.ylabel('Rata-rata Kadar pH')

# Menambahkan label di atas setiap batang
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=10)

plt.show()
```

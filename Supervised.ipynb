# Data Preparation

## Cek Outliers

```python
outliers = {}
features = gabungan_wine_df.columns
for feature in features:
    Q1 = gabungan_wine_df[feature].quantile(0.25)
    Q3 = gabungan_wine_df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menyimpan outlier
    outliers[feature] = gabungan_wine_df[(gabungan_wine_df[feature] < lower_bound) | (gabungan_wine_df[feature] > upper_bound)]

# Menampilkan jumlah outlier untuk setiap fitur
for feature, outlier_data in outliers.items():
    print(f'Jumlah outlier di {feature}: {outlier_data.shape[0]}')
```

## Menghapus Outlier

```python
# Menghapus outlier dari DataFrame
for feature in features:
    Q1 = gabungan_wine_df[feature].quantile(0.25)
    Q3 = gabungan_wine_df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    gabungan_wine_df = gabungan_wine_df[(gabungan_wine_df[feature] >= lower_bound) & (gabungan_wine_df[feature] <= upper_bound)]
```

```python
from sklearn.preprocessing import MinMaxScaler

# Memilih fitur untuk distandarisasi
features_to_scale = gabungan_wine_df.columns[:-1]  # Semua kolom kecuali 'jenis_wine' dan 'quality'

# Inisialisasi Min-Max Scaler
scaler = MinMaxScaler()

# Melakukan standarisasi
gabungan_wine_df[features_to_scale] = scaler.fit_transform(gabungan_wine_df[features_to_scale])

# Menampilkan DataFrame setelah standarisasi
print(gabungan_wine_df.head())
```

# Split Data Train Val and Test

```python
from sklearn.model_selection import train_test_split

# Misalkan gabungan_wine_df adalah DataFrame Anda
train_data, test_data = train_test_split(gabungan_wine_df, test_size=0.30, random_state=42)

# Menampilkan ukuran set
print(f'Ukuran set pelatihan: {train_data.shape[0]}')
print(f'Ukuran set pengujian: {test_data.shape[0]}')
```

## Build Model

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Mempersiapkan data
X_train = train_data.drop('quality', axis=1)
y_train = train_data['quality']

X_test = test_data.drop('quality', axis=1)
y_test = test_data['quality']

# Membangun model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Prediksi pada data pengujian
y_test_pred = model.predict(X_test)

# Evaluasi model
print(confusion_matrix(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score

# Menggunakan k-fold cross-validation
cv_scores = cross_val_score(model, X_train, y_train, cv=5)  # Menggunakan 5-fold cross-validation

print(f'Cross-validation scores: {cv_scores}')
print(f'Mean CV score: {cv_scores.mean()}')
```

## Feature Importance

```python
# Mendapatkan fitur penting
importances = model.feature_importances_
feature_names = X_train.columns

# Membuat DataFrame untuk fitur penting
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Menampilkan fitur penting
print(feature_importance_df)
```
